```python
# dividir en 2 la imagen o arreglo
x1, x2 = np.split(img_array,2)
```

```python
plt.imshow(y2)
plt.show()
```

```python
plt.imshow(np.concatenate((y1, y2), axis=1))
plt.show()
```
