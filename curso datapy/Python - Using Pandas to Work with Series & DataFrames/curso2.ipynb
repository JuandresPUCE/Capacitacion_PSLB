```python
required_cols=['Country','Population','Area (sq. mi.)','Birthrate','Deathrate',]
```

```python
paises=pd.read_csv('datasets/countries_of_the_world.csv', usecols=required_cols,decimal=',') 
paises
```

```python
#despliega la columna
paises['Country'].head()
```

```python
#valor especifico
paises['Country'][206]
```

```python
#valor en rango
paises[['Country','Birthrate']][105:112]
```

```python
#acceso a rango de columnas loc o filas
#rango de filas desde 105 a 112 no incluida
#rango de columnas desde country a birthrate

paises.loc[105:112,'Country':'Birthrate']
```

```python
#acceso a rango de columnas loc o filas
#rango de filas desde 105 a 112 no incluida
#rango de columnas desde country a birthrate
#co un paso de 3 en 3 
paises.loc[105:112:3,'Country':'Birthrate']
```

```python
#acceso a rango de columnas iloc o filas
#rango de filas desde 105 a 112 no incluida
#rango de columnas desde 0:4 iloc para no usar los headers definidos 
# #ojo 112 esta excluido

paises.iloc[105:112,0:4]
```

```python
#calculos en columna
paises['Area']=paises['Area (sq. mi.)']*1.609*1.609
```

```python
#vaciar columna de interes despues de l calculo, axis=1 columnas
paises=paises.drop(labels=['Area (sq. mi.)'],axis=1)
```

```python
#comrpobar en booleano si hay datos sin llenar
paises.isna()
```

```python
#conteo de datos no disponibles
paises.isna().sum()
```

```python
#eliminar la data faltante, es decir pierde los datos que no tengan ningun dato en birth o death rate
paises=paises.dropna()
```
